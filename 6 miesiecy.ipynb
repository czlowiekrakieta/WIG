```python
import numpy as np
import pandas as pd
import os
import tensorflow as tf

sc = "/home/lukasz/Dokumenty/python gieldowy/WIG/"

pliki = os.listdir(sc + "daneFundKlasy_za_6M")
foldery = os.listdir(sc)

bazy = ["zmiany_cen", "daneWskazniki_3M", "daneWskazniki_3-6M", "daneWskazniki_6M", "daneFundPrzeszlosc_3M", 
        "daneFundPrzeszlosc_6M", "daneFundPrzeszlosc_3-6M"]

#najl_foldery = []
zepsute = []

for spolka in pliki:
    
    najl_kwartal = ""
    najl_folder = ""
    
    try:
        for folder in bazy:
            try:
                df = pd.read_csv(sc + folder + "/" + spolka)
                df.set_index("Unnamed: 0", inplace=True)

                if df.index[0] > najl_kwartal:
                    najl_kwartal = df.index[0]
                    najl_folder = folder

                #najl_foldery.append(najl_folder)
            except:
                continue

        indeksy = pd.read_csv(sc + najl_folder + "/" + spolka).set_index("Unnamed: 0").index

        dane = pd.DataFrame( index = indeksy, data = {})


        for folder in bazy:
            df = pd.read_csv(sc + folder + "/" + spolka).set_index("Unnamed: 0")
            kolumny = []
            kolumny_w_srodku = dane.columns
            for kolumna in df.columns:
                if kolumna not in kolumny_w_srodku:
                    kolumny.append( kolumna )
            dane = pd.concat([dane, df[kolumny].loc[indeksy]], axis=1)


        klasy = pd.read_csv(sc + "daneFundKlasy_za_6M/" + spolka).set_index("Unnamed: 0")["ZKlasa"].loc[indeksy]
        klasy = pd.DataFrame( index = indeksy, data = list(klasy), columns=["ZKlasy_6M"] )
        dane = pd.concat([dane, klasy], axis=1)


        dane.drop(['ddRoznica', 'ZKlasy', 'Roznica', 'ZKlasa'], axis=1, inplace=True)
        print(spolka)
        #dane.to_csv(sc + "wszystko_za_6M/" + spolka)
    except:
        zepsute.append(spolka)
        continue

```

```python
spolki = os.listdir(sc + "wszystko_za_6M")

wszystko = pd.DataFrame(columns=dane.columns)


for spolka in spolki:
    df = pd.read_csv(sc + "wszystko_za_6M/" + spolka).set_index("Unnamed: 0")
    wszystko = pd.concat([wszystko, df], axis=0)
    
print("jeden etap")
wszystko.replace( [np.inf, -np.inf], np.nan, inplace=True)
print("drugi etap")
wszystko_beznan = wszystko[wszystko.isnull().sum(axis=1)==0]

wszystko_beznan.to_csv("wszystko_beznan_6M.csv")
```

```python
import pandas as pd
import numpy as np


branze = pd.read_csv(sc + "spolki_z_branzami.csv").set_index("Unnamed: 0").drop(["Unnamed: 0.1"], axis=1)

branze_uniq = np.unique(branze["Branza"])

kwartaly = []
for i in range(1990, 2016):
    for j in range(1,5):
        kwartaly.append(str(i) + "Q" + str(j))

        
for br in branze_uniq:
    #print(br)
    if br == 'crash':
        continue
    #print( branze[branze["Branza"]==br])
    spolki_branzowe = list(branze[branze["Branza"]==br]["Nazwa"])
    #print(spolki_branzowe)
  
    branza = pd.DataFrame( index = kwartaly, columns = wszystko_beznan.columns, dtype=np.float64 ).fillna(0)
    
    raporty = pd.DataFrame( index = kwartaly, data = {"raporty": [0]*len(kwartaly)}, dtype=np.float64)
    
    for sp in spolki_branzowe:
        if sp + ".csv" in zepsute:
            continue
        dane = pd.read_csv(sc + "wszystko_za_6M/" + sp + ".csv").set_index("Unnamed: 0")

        branza.loc[dane.index] += dane
        raporty.loc[dane.index] += 1
        
    raporty[ raporty == 0] = 1
    #print(raporty["raporty"])
    print(br)
    
    c = pd.DataFrame( {x: list(raporty["raporty"]) for x in branza.columns} )
    #print(c[-5:])
    branza = pd.DataFrame( data = branza.as_matrix()/c.as_matrix(), index = branza.index, columns = branza.columns )
    
    branza.to_csv(sc + "branze_dane/" + br + ".csv")
        
 
```
