```python
X = wszystko.iloc[np.where( wszystko.isnull().sum(axis=1) == 0)[0]]
Y = np.asarray( wszystko_klasy )[np.where( wszystko.isnull().sum(axis=1) == 0)[0] ]
```

```python
zera = np.where(Y==0)
jedynki = np.where(Y==1)

def nowe_std( tab_jedynek, tab_zer ):
    il_jedynek = len(jedynki)
    il_zer = len(zera)
    V_0 = np.var( tab_zer ) / 1e6
    V_1 = np.var( tab_jedynek ) / 1e6
    return 1e3*np.sqrt( ( il_jedynek * V_1 * il_jedynek / (il_jedynek - 1 ) + 
                     il_zer * V_0 * il_zer / ( il_zer - 1 )  )/(il_jedynek + il_zer - 2) )

def nowe_std_stab( tab_jedynek, tab_zer ):
    
    il_jedynek = len(jedynki)
    il_zer = len(zera)
    
    return np.sqrt( np.var(tab_jedynek)/il_jedynek + np.var(tab_zer)/il_zer )


tab0 = X.iloc[zera]
tab1 = X.iloc[jedynki]

Z = len(tab0)
J = len(tab1)
P = Z/(Z-1)
Q = J/(J-1)

DC = []

for wlasnosc in wlasnosci:
    
    X_z = tab0[wlasnosc]
    X_j = tab1[wlasnosc]
    
    V_0 = np.var( X_z ) 
    V_j = np.var( X_j )

    
    STD = np.sqrt(( J*V_j*Q + Z*V_0*P )/(J+Z-2))
    
    DC.append( np.absolute( np.mean(X_z) - np.mean(X_j) )/STD )
    
    
    
dcohen = pd.DataFrame( index = wlasnosci, data = DC )
dcohen.to_csv("DCOHEN_branze_3M.csv")
```
